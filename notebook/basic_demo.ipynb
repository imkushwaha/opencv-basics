```python

filepath = str(Path("../imgs/beach.jpg"))
gray_scale_img = cv2.imread(filepath, 0)
rgb_scale_img = cv2.imread(filepath, 1)
type(gray_scale_img)
```

```python
cv2.imshow("gray_scale_img", gray_scale_img)
cv2.imshow("rgb_scale_img", rgb_scale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# save image
cv2.imwrite("gray_scale_img.jpg", gray_scale_img)
```

```python
plt.imshow(rgb_scale_img)   # cv2 reads in BGR format and matplotlib uses RGB
```

## Draw the shapes

```python
rectangle_draw = cv2.rectangle(rgb_scale_img, pt1=(384,0), pt2=(510,128), color= (0,255,0), thickness=3)   #color: BGR
```

```python
cv2.imshow("rectangle_draw", rectangle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
font = cv2.FONT_ITALIC
rectangle_draw = cv2.rectangle(rgb_scale_img, pt1=(384,0), pt2=(510,128), color= (0,255,0), thickness=3)
cv2.putText(rectangle_draw, "Rectangle", (384 + 3, 0 + 30), font, 1, (255, 0, 0), 2)
cv2.imshow("rectangle_draw", rectangle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
font = cv2.FONT_ITALIC

circle_draw = cv2.circle(rgb_scale_img, center=(500,500), radius=100, color=(0,150,255), thickness=3)
cv2.imshow("circle_draw", circle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
